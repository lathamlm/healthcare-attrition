```python
# IMPORT DEPENDENCIES
from pathlib import Path
import pandas as pd
```

```python
# IMPORTING ORIGINAL FILE
attrition_og = pd.DataFrame(pd.read_csv(Path("Resources/watson_healthcare_modified.csv")))
attrition_og.head()
```

```python
# CREATES LIST OF COLUMN NAMES AND EMPTY LISTS FOR LOOP
columnname = list(attrition_og.columns)
unique_columns = []
column_attributes = []

# CREATES FOR-LOOP FOR IDENTIFYING UNIQUE VALUES PER COLUMN AND POTENTIAL PRIMARY KEYS
for column in columnname:
    numunique = attrition_og[column].nunique()
    numtotal = attrition_og[column].count()
    column_attributes.append(f"{column}, {numunique}, {numtotal}")
    if numunique == numtotal:
        unique_columns.append(column)
```

```python
# CREATES DATAFRAME FOR DATA
column_df = pd.DataFrame(column_attributes)
column_df[["Column", "Unique", "Total"]] = column_df[0].str.split(',', expand=True)
```

```python
# FORMATS DATAFRAME AND CHANGES DTYPES AS NECESSARY
column_df = column_df[["Column", "Unique", "Total"]]
column_df[["Unique", "Total"]] = column_df[["Unique", "Total"]].astype("int")
column_df.sort_values("Unique", ascending=False)
```
